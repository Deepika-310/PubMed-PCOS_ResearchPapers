```python
articlename =[]
PMID=[]
authors=[]
reference_and_year=[]
description=[]


for i in range(1,6):
    web=requests.get("https://pubmed.ncbi.nlm.nih.gov/?term=pcos%20in%20sedantary%20lifestyle&page="+str(i))
    soup=BeautifulSoup(web.text,"lxml")
    descp=soup.find_all("div",class_="full-view-snippet")
    pmidno=soup.find_all("span",class_="docsum-pmid")
    author=soup.find_all("span",class_="docsum-authors full-authors")
    refyr=soup.find_all("span",class_="docsum-journal-citation short-journal-citation")
    names=soup.find_all("a",class_="docsum-title")
    for i in descp:
        name = re.sub(r'\s+', ' ', i.text).strip()
        description.append(name)
    for i in names:
        name = re.sub(r'\s+', ' ', i.text).strip()
        articlename.append(name)
    for i in pmidno:
        name=i.text
        PMID.append(name)
    for i in refyr:
        name=i.text
        reference_and_year.append(name)
    for i in author:
        name=i.text
        authors.append(name)
    
    
```

```python
data_new=pd.DataFrame({"Article":articlename,"Authors":authors,"Ref. and Year":reference_and_year,"PMID":PMID,"Description": description})
```
